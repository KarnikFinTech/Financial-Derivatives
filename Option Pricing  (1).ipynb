```python
import numpy as np
import pandas as pd
import math as math

def price_option(optiontype, s0,k,r,u,d,n, deltat) : 
    s = np.zeros(shape=(n,n))
    c = np.zeros(shape=(n,n))
    s= pd.DataFrame(s) 
    c= pd.DataFrame(c) 
    q=0.5695   #should be changed to the formula 
    s[0][0] = s0 
    c[0][0] = s0-k
    if optiontype == 'put' :  
        for i in range(0,n-1) : 
            j=0
            for j in range(0,i+1) :   
                s[i+1][j]= u*s[i][j] 
                s[i+1][j+1] = d*s[i][j]
                if k-(u*s[i][j]) >= 0 :
                    c[i+1][j] = k-(u*s[i][j])
                else : 
                    c[i+1][j] = 0 
                if k-(d*s[i][j]) >= 0 :
                    c[i+1][j+1] = k-(d*s[i][j])
                else: 
                    c[i+1][j+1] = 0
                    
    elif optiontype == 'call' : 
        
        for i in range(0,n-1) : 
            j=0
            for j in range(0,i+1) :   
                s[i+1][j]= u*s[i][j] 
                s[i+1][j+1] = d*s[i][j]
                if (u*s[i][j])-k >= 0 :
                    c[i+1][j] = (u*s[i][j])-k
                else : 
                    c[i+1][j] = 0 
                if (d*s[i][j])-k >= 0 :
                    c[i+1][j+1] = (d*s[i][j])-k
                else: 
                    c[i+1][j+1] = 0
    else: print('please put call or put as optiontype')
        
    x = n
    for i in range(0,n-1) :
        for j in range(0,x-1) : 
            c[n-2-i][j]= (c[n-1-i][j]*q + c[n-1-i][j+1]*(1-q))*math.exp(-r*deltat)
        x -=1
                    
    print('Option price =', c[0][0])
    return(c.head(20))
    
    
```

```python
price_option('put', 100,105,0.079,1.1,0.9,100,0.083333)
```
